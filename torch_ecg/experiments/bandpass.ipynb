```python
import numpy as np

import torch
import matplotlib.pyplot as plt

from torch_ecg.cfg import DEFAULTS
from torch_ecg.preprocessors import BandPass


```

```python
# sig = DEFAULTS.RNG.randn(1000)
#sig = torch.as_tensor(np.random.randn(1000)).unsqueeze(0)
sig = np.random.randn(1000)
plt.plot(sig)
```

```python
sig_t = torch.as_tensor(sig).unsqueeze(0)
```

```python
pp = BandPass(fs=500, lowcut=0.5, highcut=45, filter_type="butter", filter_order=4)
#pp = BandPass(fs=500, lowcut=1.0, highcut=100, filter_type="butter", filter_order=4)
out_sig = pp(sig_t)
```
